```python
import PIL
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLO model
model = YOLO('weights/palmoil_10.pt')

# Load the image
source_img = 'images/palm.jpg'
uploaded_image = PIL.Image.open(source_img)

# Set the confidence threshold
confidence = 0.4

# Make predictions
res = model.predict(uploaded_image, conf=confidence)

# Get the bounding boxes
boxes = res[0].boxes

# Plot the results
res_plotted = res[0].plot()[:, :, ::-1]

# Display the image with matplotlib
plt.imshow(res_plotted)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import torch
from PIL import Image
import matplotlib.pyplot as plt

# Load YOLOv7 model
model = torch.hub.load('WongKinYiu/yolov7', 'custom', 'weights/best.pt', source='local')

# Load the image
source_img = 'images/office_4.jpg'
uploaded_image = Image.open(source_img)

# Perform inference
results = model(uploaded_image, size=640)

# Print results
results.print()

# Display results
results.show()

# Save results
results.save()
```
